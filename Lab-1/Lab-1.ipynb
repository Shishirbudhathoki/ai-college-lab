### Lab 1: Problem Solving by Searching: N Queens Problem

```python
def is_safe(board, x, y, n):

    #Column Test
    for row in range(x):
        if board[row][y] == 'Q':
            return False
        
    # Left Upper Diagonal Test
    r = x - 1
    c = y - 1
    while(r>=0 and c>=0 ):
            if board[r][c] == 'Q':
                return False
            r = r - 1
            c = c - 1

    r = x - 1
    c = y + 1
    while(r>=0 and c<n ):
            if board[r][c] == 'Q':
                return False
            r = r - 1
            c = c + 1
    return True

        

```

```python
from pprint import pprint
n = 4
board = [['_']*n for i in range(n)]
board[0][3] = 'Q'
pprint(board)
print(is_safe(board=board,x=1,y=2,n=n))
```

```python
from pprint import pprint
n = int(input("Enter number of Q's "))
board = [['_']*n for i in range(n)]
def nQueenSolver(board,row,n):
    if(row>=n):
        return True
    for col in range(n):
        if is_safe(board,row,col,n):
            board[row][col] = 'Q'
            if(nQueenSolver(board,row+1,n)):
                return True
            board[row][col] = "_"
    return False

if(nQueenSolver(board,0,n)):
    pprint(board)
else:
    pprint("NO SOLUTION")
```
