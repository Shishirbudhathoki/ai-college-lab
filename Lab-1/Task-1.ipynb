## Task 1: Goat, Wolf and Cabbage Problem


```python
def is_valid(state):
    F, W, G, C = state

    # Wolf and Goat together without farmer
    if W == G and F != W:
        return False

    # Goat and Cabbage together without farmer
    if G == C and F != G:
        return False

    return True
```

```python
# Generate next states
def generate_next_states(state):
    F, W, G, C = state
    next_states = []

    # Farmer alone
    new_state = (1-F, W, G, C)
    if is_valid(new_state):
        next_states.append(new_state)

    # Farmer + Wolf
    if F == W:
        new_state = (1-F, 1-W, G, C)
        if is_valid(new_state):
            next_states.append(new_state)

    # Farmer + Goat
    if F == G:
        new_state = (1-F, W, 1-G, C)
        if is_valid(new_state):
            next_states.append(new_state)

    # Farmer + Cabbage
    if F == C:
        new_state = (1-F, W, G, 1-C)
        if is_valid(new_state):
            next_states.append(new_state)

    return next_states
```

```python
from collections import deque

def river_crossing():
    start = (0,0,0,0)
    goal = (1,1,1,1)

    queue = deque([(start,[start])])
    visited = set()

    while queue:
        state, path = queue.popleft()
        if state == goal:
            print(path)
            return path
        visited.add(state)
        for next_state in generate_next_states(state):
            if next_state not in visited:
                queue.append((next_state,path+[next_state]))

    return None

solution = river_crossing()
print("Solution Path:\n")
for s in solution:
    print(s)
```

```python
def print_solution(path):
    names = ["Farmer", "Wolf", "Goat", "Cabbage"]

    for step, state in enumerate(path):
        west = []
        east = []

        for i in range(4):
            if state[i] == 0:
                west.append(names[i])
            else:
                east.append(names[i])

        print(f"Step {step}")
        print("West Side :", west)
        print("East Side :", east)
        print("-" * 30)

print_solution(solution)
```
