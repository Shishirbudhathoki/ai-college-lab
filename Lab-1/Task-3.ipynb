## Task 3: Robot Navigation Problem

```python
from collections import deque

grid = [
    ['S', '.', '.', 'X', '.'],
    ['.', 'X', '.', 'X', '.'],
    ['.', 'X', '.', '.', '.'],
    ['.', '.', 'X', 'X', '.'],
    ['.', '.', '.', '.', 'G']
]

ROWS = 5
COLS = 5

start = (0, 0)
goal = (4, 4)
```

```python

def bfs():
    queue = deque([(start, [start])])
    visited = set()

    directions = [
        (-1, 0),  # Up
        (1, 0),   # Down
        (0, -1),  # Left
        (0, 1)    # Right
    ]

    while queue:
        current, path = queue.popleft()

        if current == goal:
            return path

        visited.add(current)

        row, col = current

        for dr, dc in directions:
            nr = row + dr
            nc = col + dc

            next_state = (nr, nc)

            if (0 <= nr < ROWS and 0 <= nc < COLS and grid[nr][nc] != 'X' and next_state not in visited):
                queue.append((next_state, path + [next_state]))

    return None

solution = bfs()

print("Shortest Path:")
for state in solution:
    print(state)
```

```python
def print_solution(path):
    for step, state in enumerate(path):
        row, col = state

        print(f"Step {step}")
        print(f"Robot Position: ({row}, {col})")
        print("-" * 30)

print_solution(solution)
```
